```python
def calculate_metrics_no_lib(simulated, real):
    """
    Calculate p-value, ME, MAE, RMSE, MAPE, and RMSPE without using libraries.
    """
    # Ensure inputs are lists for manual calculations
    simulated = list(simulated)
    real = list(real)

    # Calculate Mean Error (ME)
    n = len(simulated)
    
    # show the mae calculation as cummulative sum of the difference between simulated and real
    print()
    cum_sum = 0
    for sim, re in zip(simulated, real):
        cum_sum += sim - re
        print(sim, re, sim-re, cum_sum)
    print(f"me found {cum_sum/n}")
    
    me = sum(sim - re for sim, re in zip(simulated, real)) / n

    # Calculate Mean Absolute Error (MAE)
    mae = sum(abs(sim - re) for sim, re in zip(simulated, real)) / n

    # Calculate Root Mean Square Error (RMSE)
    mse = sum((sim - re) ** 2 for sim, re in zip(simulated, real)) / n
    rmse = mse ** 0.5

    # Calculate Mean Absolute Percentage Error (MAPE)
    mape = sum(abs((sim - re) / re) for sim, re in zip(simulated, real)) / n * 100

    # Calculate Root Mean Square Percentage Error (RMSPE)
    mspe = sum(((sim - re) / re) ** 2 for sim, re in zip(simulated, real)) / n
    rmspe = (mspe ** 0.5) * 100

    # Calculate p-value using library
    p_value = ttest_ind(simulated, real).pvalue
    

    return p_value, me, mae, rmse, mape, rmspe

# Prepare DataFrame for results
columns = ["Density", "Metric", "p-value", "ME", "MAE", "RMSE", "MAPE", "RMSPE"]
results = []

# For SP
real_sp_car_high = real_time_avg_high  # Replace with actual SP real-time data for car
real_sp_bike_high = real_time_avg_high_bike  # Replace with actual SP real-time data for bike
real_sp_car_moderate = real_time_avg_moderate  # Replace with actual SP real-time data for car
real_sp_bike_moderate = real_time_avg_moderate_bike  # Replace with actual SP real-time data for bike


# Moderate Density - SP
print("now calculating for sp car moderate")
metrics = calculate_metrics_no_lib(sp_car_moderate_with, [real_sp_car_moderate] * len(sp_car_moderate_with))
results.append(["Moderate", "SP Car", *metrics])
print("now calculating for sp bike moderate")
metrics = calculate_metrics_no_lib(sp_bike_moderate_with, [real_sp_bike_moderate] * len(sp_bike_moderate_with))
results.append(["Moderate", "SP Bike", *metrics])

# High Density - SP
print("now calculating for sp car high")
metrics = calculate_metrics_no_lib(sp_car_high_with, [real_sp_car_high] * len(sp_car_high_with))
results.append(["High", "SP Car", *metrics])
print("now calculating for sp bike high")
metrics = calculate_metrics_no_lib(sp_bike_high_with, [real_sp_bike_high] * len(sp_bike_high_with))
results.append(["High", "SP Bike", *metrics])

# For PS
real_ps_car_high = real_time_avg_high_car_ps  # Replace with actual PS real-time data for car
real_ps_bike_high = real_time_avg_high_bike_ps  # Replace with actual PS real-time data for bike
real_ps_car_moderate = real_time_avg_moderate_car_ps  # Replace with actual PS real-time data for car
real_ps_bike_moderate = real_time_avg_moderate_bike_ps  # Replace with actual PS real-time data for bike

# Moderate Density - PS
print("now calculating for ps car moderate")
metrics = calculate_metrics_no_lib(ps_car_moderate_with, [real_ps_car_moderate] * len(ps_car_moderate_with))
results.append(["Moderate", "PS Car", *metrics])
print("now calculating for ps bike moderate")
metrics = calculate_metrics_no_lib(ps_bike_moderate_with, [real_ps_bike_moderate] * len(ps_bike_moderate_with))
results.append(["Moderate", "PS Bike", *metrics])

# High Density - PS
print("now calculating for ps car high")
metrics = calculate_metrics_no_lib(ps_car_high_with, [real_ps_car_high] * len(ps_car_high_with))
results.append(["High", "PS Car", *metrics])
print("now calculating for ps bike high")
metrics = calculate_metrics_no_lib(ps_bike_high_with, [real_ps_bike_high] * len(ps_bike_high_with))
results.append(["High", "PS Bike", *metrics])

# Convert results to a DataFrame
metrics_df = pd.DataFrame(results, columns=columns)

# Display results
print(metrics_df)
```
