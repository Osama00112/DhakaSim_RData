```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
data = pd.read_csv("file4.csv")
```

```python
# Define the desired sorting order of the x-axis labels
x_order = ["C4-{}".format(i) for i in range(1, 5)]

# Convert the "Case Code" column into a categorical data type with the desired sorting order
data["Case Code"] = pd.Categorical(data["Case Code"], categories=x_order)

# Group by case code and average speed
grouped = data.groupby(["Case Code"])["Avg. Speed"].mean()

# Set font family to Times New Roman for all text elements
matplotlib.rcParams['font.family'] = 'Times New Roman'

# Plot bar graph
grouped.plot(kind="bar", zorder = 2, color = cm.Oranges(0.7))
#plt.ylim(8.30,9.30)
plt.ylabel("Avg. Speed(km/h)", fontsize=18, labelpad=10)  # increase font size
plt.xlabel("Case No.", fontsize=18, labelpad=10, fontname = "Times New Roman")  # increase font size
plt.tick_params(labelsize=16)  # increase tick font size
plt.grid(True, axis="y", color = "black", linestyle = "--")
for i in range(len(grouped)):
    plt.axvline(x=i - 0.5, color="black",  alpha=0.5, linestyle = "--")


# Adjust spacing between tick labels and axes


plt.savefig("plots/avg1-edited2.png", dpi=300, bbox_inches='tight')
plt.show()
```
