```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("file.csv")
```

```python
# Group by case code and average speed
grouped = data.groupby(["Case Code"])["Avg. Waiting Time"].mean()

# Plot bar graph
grouped.plot(kind="bar", zorder = 2)
plt.ylim(7.5,10)
plt.ylabel("Avg. Waiting Time(sec)", fontsize=16, labelpad=10)  # increase font size
plt.xlabel("Case No.", fontsize=16, labelpad=10)  # increase font size
plt.tick_params(labelsize=14)  # increase tick font size
plt.grid(True, axis="y", color = "black", linestyle = "--")
for i in range(len(grouped)):
    plt.axvline(x=i - 0.5, color="black",  alpha=0.5, linestyle = "--")

# Adjust spacing between tick labels and axes
plt.savefig("plots/avg waiting time.png", dpi=300, bbox_inches='tight')
plt.show()
```
