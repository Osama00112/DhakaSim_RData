```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("file2.csv")
```

```python
# Define the desired sorting order of the x-axis labels
x_order = ["C1-{}".format(i) for i in range(1, 16)]

# Convert the "Case Code" column into a categorical data type with the desired sorting order
data["Case Code"] = pd.Categorical(data["Case Code"], categories=x_order)

# Group by case code and average speed
grouped = data.groupby(["Case Code"])["Avg. Waiting Time"].mean()
# Plot bar graph
grouped.plot(kind="bar", zorder = 2, color = "brown")
#plt.ylim(7.5,10)
plt.ylabel("Avg. Waiting Time(sec)", fontsize=20, labelpad=10, fontname = "Times New Roman")  # increase font size
plt.xlabel("Case No.", fontsize=20, labelpad=10, fontname = "Times New Roman")  # increase font size
plt.tick_params(labelsize=14)  # increase tick font size
plt.grid(True, axis="y", color = "black", linestyle = "--")
for i in range(len(grouped)):
    plt.axvline(x=i - 0.5, color="black",  alpha=0.5, linestyle = "--")

# Adjust spacing between tick labels and axes
plt.savefig("plots/avg7-edited2.png", dpi=300, bbox_inches='tight')
plt.show()
```
